## ESOL solubility dataset, `pandas`, `SQL`, `rdkit`


The ESOL dataset (*Delaney 2004*) is a publicly available set of measured solubilities for an agricultural set of 1128 small molecules.  First take a look with Pandas

```python
import pandas as pd
df = pd.read_csv('CDT_TPT_data/delaney_raw.csv')
df
```

Neaten the column names for SQL insertion. SQL minimizes storage space and allows quick retrieval. It enforces discipline in data management which can be helpful.

```python
# setup the sqlite connection
import sqlite3
conn = sqlite3.connect('results.db')
# rename table columns from the raw data
original_cols = df.columns.copy() 
df.columns = ['name', 'ESOL_pred', 'min_deg', 'mol_wt', 'N_rings', 'N_HBDs', 'N_RBS', 'SA', 'ESOL_true', 'SMILES']
df.to_sql('esol', conn)
df
```

Our new nams help in sql queries, but they are not easy to read. We can store a key of column names in the database too.

```python
c = conn.cursor()
c.execute('CREATE TABLE IF NOT EXISTS column_key (tbl TEXT, col_name TEXT, definition TEXT UNIQUE)')
conn.commit()

table = 'esol'
for tup in zip(df.columns, original_cols):
  print(tup)
  # the ? is replaced by tuple values. 
  # the 'OR IGNORE' means the process won't fail if the unique constraint is voilated
    # '*' 'unpacks the tuple
  c.execute('INSERT OR IGNORE INTO column_key (tbl, col_name, definition) VALUES (?, ?, ?)', (table, *tup)) 
conn.commit()
```

In this example, make a table and generate extra features.

```python
import random
import rdkit
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw

# seed the random
random.seed(90)
# pick a random dataframe index
idx = random.randrange(df.shape[0])
# get the smiles
name, mw, smiles = tuple(df.loc[idx, ['name', 'mol_wt', 'SMILES']])
print(f'{name}, mol wt = {mw:.3f}')
mol = Chem.MolFromSmiles(smiles)
mol
```

```python
from rdkit.Chem import Crippen
from rdkit.Chem import inchi

extended_esol_data = lambda s: (s, 
                                inchi.MolToInchiKey(Chem.MolFromSmiles(s)), 
                                Crippen.MolLogP(Chem.MolFromSmiles(s)))
extended_esol_data(smiles)
```

```python
# make the new table
c.execute('CREATE TABLE IF NOT EXISTS ex_esol (SMILES_ex TEXT UNIQUE, InChiKey TEXT UNIQUE, Crippen_logP REAL)')
# set an extra cursor
c2 = conn.cursor()
for row in c2.execute('SELECT SMILES, SA FROM esol'):
  c.execute('INSERT OR IGNORE INTO ex_esol (SMILES_ex, InChiKey, Crippen_logP) VALUES (?,?,?)', extended_esol_data(row[0]))
conn.commit()

# then we can view the joined table.
pd.read_sql("""SELECT SMILES, Crippen_logP, min_deg, mol_wt, N_rings, N_HBDs, SA, ESOL_pred, ESOL_true FROM esol 
          INNER JOIN ex_esol 
          ON esol.SMILES = ex_esol.SMILES_ex""", conn)
```

We can save the above as a view

```python
c.execute("""CREATE VIEW esol_logP AS 
          SELECT SMILES, Crippen_logP, min_deg, mol_wt, N_rings, N_HBDs, SA, ESOL_pred, ESOL_true FROM esol 
          INNER JOIN ex_esol 
          ON esol.SMILES = ex_esol.SMILES_ex""")
conn.commit()
# then you can read the view directly
pd.read_sql('SELECT * FROM esol_logP', conn)
```

`sqlite_master` holds a list of tables. We can define its column names too.

```python
# with a dict
sqlite_master_colnames = {'type' : 'Type is Index or Table.',
                            'name' : 'Index or table name',
                            'tbl_name' : 'Name of the table',
                            'rootpage' : 'rootpage, internal to SQLite',
                            'sql' : 'The SQL statement, which would create this table'}

table = 'sqlite_master'
for col_name, definition in sqlite_master_colnames.items(): 
  c.execute('INSERT INTO column_key (tbl, col_name, definition) VALUES (?, ?, ?)', 
            (table, col_name, definition))
conn.commit()

pd.read_sql('SELECT * FROM column_key', conn)
```

The sql pragma function lets you read the table info

```python
pd.read_sql("""PRAGMA table_info(esol)""", conn)
```

### Preprocessing

We'll split the dataset into 10 sections. 90% of the data are fed into model, the remainder is used to evaluate the test. 
There are special splitting techniques, but this will be a random shuffle. 

```python
df = pd.read_sql('SELECT * FROM esol_logP', conn)
seed = 5
random.seed(seed)
n_entries = len(df.index)
slices = [(round(i*n_entries/10), round((i+1)*n_entries/10)) for i in range(10)]
print(slices)
shuffled_indices = random.sample(list(df.index), n_entries)
splits = [[(j,i) for j in shuffled_indices[sls[0]: sls[1]]] for i, sls in enumerate(slices)]
splits = []
for i, sls in enumerate(slices):
  for j in shuffled_indices[sls[0]: sls[1]]:
    splits.append([j,i])
splits_df = pd.DataFrame(splits, columns=['idx', 'split'])
splits_df
```

```python
# write the splits to the datbase
splits_df.to_sql(f'splits_{seed}', conn, index=False, if_exists='replace')
c.execute("""CREATE VIEW IF NOT EXISTS split_esol AS SELECT SMILES, Crippen_logP, min_deg, mol_wt, N_rings, N_HBDs, SA, ESOL_pred, ESOL_true, split
              FROM esol 
              JOIN ex_esol ON esol.SMILES = ex_esol.SMILES_ex
              INNER JOIN splits_5 ON esol.rowid = splits_5.idx""")
conn.commit()
df = pd.read_sql('SELECT * FROM split_esol', conn)
df
```

### MLR

```python
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression

idf = pd.read_sql(f"""SELECT split, Crippen_logP,	min_deg,	mol_wt,	N_rings,	N_HBDs,	SA, ESOL_true
                            FROM split_esol""", conn)

rdf = pd.DataFrame()
ndf =  idf[['split', 'ESOL_true']].copy()
for i in range(10):
  train = idf[idf['split'] != i]
  test = idf[idf['split'] == i]
  xy = lambda df: (df[['Crippen_logP',	'min_deg',	'mol_wt',	'N_rings',	'N_HBDs',	'SA']], df['ESOL_true'])
  train_x, train_y = xy(train)
  test_x, test_y = xy(test)
  X = df[['Crippen_logP',	'min_deg',	'mol_wt',	'N_rings',	'N_HBDs',	'SA']]
  mlr = sklearn.linear_model.LinearRegression()
  mlr.fit(train_x, train_y)
  data = {'split': i}
  data["Intercept"] = mlr.intercept_
  for label, coeff in zip(train_x, mlr.coef_):
    data[label] = coeff
  data['train_rmse'] = np.sqrt(np.mean(np.square(mlr.predict(train_x) - train_y))) # rmse
  data['test_rmse'] = np.sqrt(np.mean(np.square(mlr.predict(test_x) - test_y)))
  rdf = rdf.append(data, ignore_index=True)
  # ns.loc[idf[idf['split'] != i].index] = list(mlr.predict(train_x))
  # ns.loc[idf[idf['split'] == i].index] = list(mlr.predict(test_x))
  ndf[f'split_{i}'] = mlr.predict(X)

print(ndf)
print(rdf)
ndf.to_sql('mlr_split_results', conn, if_exists='replace')
rdf.to_sql('mlr_fits', conn, if_exists='replace')
```
